```python
import csv
from matplotlib import pyplot


variete = []
sew = []
pew = []

with open("iris.csv",mode='r',encoding='utf8',newline='') as f:
    data = csv.reader(f,delimiter=',')
    for row in data:
        if row[0] != 'ID':
            variete.append(row[0][5:])
            sew.append(int(row[2]))
            pew.append(int(row[4]))
```

```python
# Paramétrage des axes
axes = pyplot.gca()
axes.set_xlabel('Largeur des sépales')
axes.set_ylabel("Largeur des pétales")

for i in range(150):
    if variete[i]=='setosa':
        color='orange'
    elif variete[i]=='virginica':
        color='blue'
    else:
        color='green'
    pyplot.scatter(sew[i], pew[i],c=color)
pyplot.scatter(35,2,c='black')
pyplot.scatter(30,14,c='black')
pyplot.scatter(30,18,c='black')
pyplot.show()
```

```python
from math import sqrt

k=round(sqrt(150))
print(k)
```

```python
def distance(xA,yA,xB,yB):
    return sqrt((xA-xB)**2+(yA-yB)**2)
```

```python
base=[]
for i in range(150):
    iris=[sew[i],pew[i],variete[i]]
    base.append(iris)
    
    
def kNN(liste,x,y,k):
    # on crée une liste vide qui contient des tuples composés de la distance entre le nouvel iris 
    # et un iris de la base ainsi que sa classe (distance, classe)
    distance_variete = []
    # on clacule les distances entre le nouvel iris et les iris de la base
    # on ajoute les tuples dans la liste
    for i in range(150):
        distance_variete.append((distance(liste[i][0],liste[i][1],x,y),liste[i][2]))
        
    # on trie la liste des tuples selon la distance
    distance_variete.sort()
    
    # on recherche la classe majoritaire parmi les k premiers tuples da la liste triée distance_variete
    # on crée une novelle liste qui va contenir pour chaque classe le nombre d'occurences
    varietes = []
    
    # on parcourt les k premiers tuples et compte les occurences de classe
    for i in range(k):
        varietes.append(distance_variete[i][1])
        
    # la liste compteur_variete va rassembler dans des tuples chaque classe et son nombre d'occurences
    compteur_variete = []
    for v in varietes:
        if (varietes.count(v),v) not in compteur_variete:
            compteur_variete.append((varietes.count(v),v))
    
    # on renvoie la classe qui a la plus forte occurence.
    return max(compteur_variete)[1]
        
```

```python
kNN(base,35,2,k)
```
