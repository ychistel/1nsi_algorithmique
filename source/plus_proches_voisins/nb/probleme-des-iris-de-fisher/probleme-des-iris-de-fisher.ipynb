```python
import csv
from matplotlib import pyplot

variete = []
sew = []
pew = []

with open("iris.csv",mode='r',encoding='utf8',newline='') as f:
    data = csv.reader(f,delimiter=',')
    for fleur in data:
        #print(fleur)
        if fleur[0] != 'ID':
            variete.append(fleur[0][5:])
            sew.append(int(fleur[2]))
            pew.append(int(fleur[-1]))
```

```python
for i in range(150):
    if variete[i] == 'setosa':
        pyplot.scatter(sew[i],pew[i],c='orange')
    elif variete[i] == 'virginica':
        pyplot.scatter(sew[i],pew[i],c='blue')
    else:
        pyplot.scatter(sew[i],pew[i],c='green')
pyplot.show()
```
