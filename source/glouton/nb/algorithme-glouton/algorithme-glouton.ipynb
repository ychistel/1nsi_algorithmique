```python
def rendu_monnaie(somme_a_rendre):
    """somme à rendre (int) -> nombre de pièces (int)"""
    pieces = [200,100,50,20,10,5,2,1]
    n = 0 # nombre pièces
    for p in pieces:
        while somme_a_rendre >= p:
            somme_a_rendre = somme_a_rendre - p
            n = n + 1
    return n
```

```python
# La fonction renvoie les pièces et billets utilisés

def rendu_monnaie(somme_a_rendre):
    # valeurs pièces et billets utilisés
    tableau_des_unites = [200,100,50,20,10,5,2,1]
    unites = []
    i = 0
    piece = tableau_des_unites[i]
    while somme_a_rendre > 0:
        if somme_a_rendre >= piece:
            somme_a_rendre = somme_a_rendre - piece
            unites.append(piece)
        else:
            i = i + 1
            piece = tableau_des_unites[i]
    return unites
```

```python
def rendu_monnaie(somme_a_rendre):
    """somme à rendre (int) -> nombre de pièces (int)"""
    pieces = [200,100,50,20,10,5,2,1]
    unites = [] # nombre pièces
    for p in pieces:
        while somme_a_rendre >= p:
            somme_a_rendre = somme_a_rendre - p
            unites.append(p)
    return unites
```
